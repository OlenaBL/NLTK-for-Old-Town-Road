1) Tokenizing

```python
import nltk
from nltk.tokenize import word_tokenize
my_data_tokens = word_tokenize(my_data)
my_data_tokens
print(my_data_tokens)
print (len(my_data_tokens))
```

2) Removing punctuation

```python
from nltk.tokenize import RegexpTokenizer
my_tokenizer = RegexpTokenizer(r'\w+')
my_tokenized_list = my_tokenizer.tokenize(my_data)
print(my_tokenized_list)
```

3) Removing stop words (English)

```python
from nltk.corpus import stopwords
def remove_stopwords(my_tokenized_list):
    new_data = [word for word in my_tokenized_list if word not in stopwords.words('english')]
    return new_data
new_data = remove_stopwords(my_tokenized_list)
print(new_data)
```

4) Stemming and Lemmatizing (comparison)

```python
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
ps = PorterStemmer()
lemmatizer = WordNetLemmatizer()

def stemmed_data(new_data):
    my_stemmed_list = [ps.stem(word) for word in new_data]
    return my_stemmed_list
my_stemmed_list = stemmed_data(new_data)
print(my_stemmed_list)

def lemmatized_data(new_data):
    my_lemmatized_list = [lemmatizer.lemmatize(word) for word in new_data]
    return my_lemmatized_list
my_lemmatized_list = lemmatized_data(new_data)
print(my_lemmatized_list)
```

5) Frequency distribution (with a graph)

```python
from nltk import FreqDist

freq_dist = FreqDist(my_lemmatized_list)
print(freq_dist)
print(list(freq_dist.keys()))

freq_dist.plot(10)
freq_dist.plot(20)
```
